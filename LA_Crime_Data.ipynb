```python
# Import Libraries
import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("sumaiaparveenshupti/los-angeles-crime-data-20102020")
print("Path to dataset files:", path)
```

```python
# List all files in the specified directory and store the file names in the `files` list
files = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
print("Files in directory:", files)
```

```python
# Load the first CSV file into a DataFrame
la_crime_data_2010_to_2019 = pd.read_csv(os.path.join(path, files[0]))
la_crime_data_2010_to_2019
```

```python
# Load the second CSV file into a DataFrame
la_crime_data_2020_to_2021 = pd.read_csv(os.path.join(path, files[1]))
la_crime_data_2020_to_2021
```

```python
# Combine datasets
la_crime_data_2010_to_2021 = pd.concat([la_crime_data_2010_to_2019, la_crime_data_2020_to_2021], ignore_index=True)
la_crime_data_2010_to_2021 
```

```python
# Inspect column names
la_crime_data_2010_to_2021.columns.tolist()
```

```python
# Removed unnecessary spacing
la_crime_data_2010_to_2021.columns = la_crime_data_2010_to_2021.columns.str.strip()

# Dropped crime code columns 2, 3, 4
la_crime_data_2010_to_2021 = la_crime_data_2010_to_2021.drop(columns=['Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4'])
la_crime_data_2010_to_2021
```

```python
# Remove duplicate rows from the DataFrame
la_crime_data_2010_to_2021 = la_crime_data_2010_to_2021.drop_duplicates()
la_crime_data_2010_to_2021 
```

```python
# Fill NaNs with 0
la_crime_data_2010_to_2021.fillna(0)
```

```python
# Check the data types of each column
print(la_crime_data_2010_to_2021.dtypes)
```

```python
# Convert LAT and LON columns to geometry tuples
la_crime_data_2010_to_2021['Geometry'] = la_crime_data_2010_to_2021.apply(
    lambda row: (row['LON'], row['LAT']), axis=1
)
# Move the geometry column to be right after longitude
lon_index = la_crime_data_2010_to_2021.columns.get_loc('LON')
la_crime_data_2010_to_2021.insert(lon_index + 1, 'Geometry', la_crime_data_2010_to_2021.pop('Geometry'))
la_crime_data_2010_to_2021
```

```python
la_crime_data_2010_to_2021
```

```python
# Convert 'Date Rptd' and 'DATE OCC' columns from string to datetime objects to enable easier date and time manipulation.
la_crime_data_2010_to_2021['Date Rptd']  = pd.to_datetime(la_crime_data_2010_to_2021['Date Rptd'] , format='%m/%d/%Y %I:%M:%S %p')
la_crime_data_2010_to_2021['DATE OCC'] = pd.to_datetime(la_crime_data_2010_to_2021['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
# Sort the DataFrame based on 'DATE OCC'
la_crime_data_2010_to_2021_sorted = la_crime_data_2010_to_2021.sort_values(by='DATE OCC')
la_crime_data_2010_to_2021_sorted 
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define age groups
def categorize_age(age):
    if age < 18:
        return 'Under 18'
    elif 18 <= age < 30:
        return '18-29'
    elif 30 <= age < 50:
        return '30-49'
    elif 50 <= age < 65:
        return '50-64'
    else:
        return '65+'

# Categorize age if not already done
if 'Age Group' not in la_crime_data_2010_to_2021_sorted.columns:
    la_crime_data_2010_to_2021_sorted['Age Group'] = la_crime_data_2010_to_2021_sorted['Vict Age'].apply(categorize_age)

# Count crimes by age group
age_group_counts = la_crime_data_2010_to_2021_sorted['Age Group'].value_counts()

# Plotting a pie chart for age groups involved in crimes
plt.figure(figsize=(10, 8))
age_group_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired(range(len(age_group_counts))))
plt.title('Distribution of Crimes by Age Group')
plt.ylabel('')  # Hide y-axis label for cleanliness
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert 'TIME OCC' to hour of the day
la_crime_data_2010_to_2021_sorted['Hour'] = la_crime_data_2010_to_2021_sorted['TIME OCC'] // 100

# Group by hour and age group
hourly_crimes_by_age = la_crime_data_2010_to_2021_sorted.groupby(['Hour', 'Age Group']).size().unstack(fill_value=0)

# Adding a total crimes per hour line
hourly_crimes_by_age['Total'] = hourly_crimes_by_age.sum(axis=1)

# Plotting crimes per hour, colored by age groups and including total
plt.figure(figsize=(14, 7))
for column in hourly_crimes_by_age.columns:
    plt.plot(hourly_crimes_by_age.index, hourly_crimes_by_age[column], label=column)

plt.title('Crimes Per Hour by Age Group and Total')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crimes')
plt.legend(title='Age Group / Total')
plt.grid(True)
plt.xticks(range(0, 24))  # Ensure all hours are shown
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Group by 'Age Group' and 'Crm Cd Desc', then count occurrences
crime_counts = la_crime_data_2010_to_2021_sorted.groupby(['Age Group', 'Crm Cd Desc'], observed=True).size()

# Sort within each age group and select the top 5 crimes
top_crimes_by_group = (
    crime_counts.groupby(level=0, group_keys=False, observed=True)
    .nlargest(5)
    .reset_index(name='Count')
)

# Pivot the table so that age groups are columns and ranks are rows
ranked_table = (
    top_crimes_by_group.assign(Rank=top_crimes_by_group.groupby('Age Group').cumcount() + 1)
    .pivot(index='Rank', columns='Age Group', values='Crm Cd Desc')
)

# Display the table
print("Top 5 Crimes by Age Group (Transposed Table):")
print(ranked_table)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Mapping descent codes to full descriptions
descent_mapping = {
    'A': 'Other Asian', 'B': 'Black', 'C': 'Chinese', 'D': 'Cambodian',
    'F': 'Filipino', 'G': 'Guamanian', 'H': 'Hispanic/Latin/Mexican',
    'I': 'American Indian/Alaskan Native', 'J': 'Japanese', 'K': 'Korean',
    'L': 'Laotian', 'O': 'Other', 'P': 'Pacific Islander', 'S': 'Samoan',
    'U': 'Hawaiian', 'V': 'Vietnamese', 'W': 'White', 'X': 'Unknown',
    'Z': 'Asian Indian'
}

# Assuming 'la_crime_data_2010_to_2021_sorted' is your DataFrame
la_crime_data_2010_to_2021_sorted['Date Rptd'] = pd.to_datetime(la_crime_data_2010_to_2021_sorted['Date Rptd'])
la_crime_data_2010_to_2021_sorted['DATE OCC'] = pd.to_datetime(la_crime_data_2010_to_2021_sorted['DATE OCC'])
la_crime_data_2010_to_2021_sorted['Report Delay'] = (la_crime_data_2010_to_2021_sorted['Date Rptd'] - la_crime_data_2010_to_2021_sorted['DATE OCC']).dt.days

# Filter and group by 'Vict Descent'
valid_descents = ['A', 'B', 'C', 'D', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'O', 'P', 'S', 'U', 'V', 'W', 'X', 'Z']
filtered_data = la_crime_data_2010_to_2021_sorted[la_crime_data_2010_to_2021_sorted['Vict Descent'].isin(valid_descents)]
stats_by_descent = filtered_data.groupby('Vict Descent').agg({
    'Report Delay': 'mean',
    'Crm Cd': 'mean'
}).reset_index()

# Normalize Crime Seriousness and sort by Report Delay
stats_by_descent['Normalized Crime Seriousness'] = (stats_by_descent['Crm Cd'] - stats_by_descent['Crm Cd'].min()) / (stats_by_descent['Crm Cd'].max() - stats_by_descent['Crm Cd'].min())
stats_by_descent.sort_values('Report Delay', ascending=False, inplace=True)
stats_by_descent['Descent Description'] = stats_by_descent['Vict Descent'].map(descent_mapping)

# Visualization
fig, ax = plt.subplots(figsize=(14, 8))
colors = [plt.cm.Reds(x) for x in stats_by_descent['Normalized Crime Seriousness']]
bars = ax.bar(stats_by_descent['Descent Description'], stats_by_descent['Report Delay'], color=colors)
ax.set_xlabel('Victim Descent')
ax.set_ylabel('Average Report Delay (days)')
ax.set_title('Average Reporting Delay by Victim Descent with Crime Seriousness')
cbar = fig.colorbar(plt.cm.ScalarMappable(norm=plt.Normalize(vmin=0, vmax=1), cmap='Reds'), ax=ax, label='Crime Seriousness (Normalized)')

# Set x-ticks and align text
ax.set_xticklabels(stats_by_descent['Descent Description'], rotation=45, ha='right', rotation_mode='anchor')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Mapping descent codes to full descriptions
descent_mapping = {
    'A': 'Other Asian', 'B': 'Black', 'C': 'Chinese', 'D': 'Cambodian',
    'F': 'Filipino', 'G': 'Guamanian', 'H': 'Hispanic/Latin/Mexican',
    'I': 'American Indian/Alaskan Native', 'J': 'Japanese', 'K': 'Korean',
    'L': 'Laotian', 'O': 'Other', 'P': 'Pacific Islander', 'S': 'Samoan',
    'U': 'Hawaiian', 'V': 'Vietnamese', 'W': 'White', 'X': 'Unknown',
    'Z': 'Asian Indian'
}

# Calculate the total cases by victim descent
cases_by_descent = la_crime_data_2010_to_2021_sorted['Vict Descent'].value_counts()

# Apply descent mapping to index
cases_by_descent.index = [descent_mapping.get(idx, idx) for idx in cases_by_descent.index]

# Group smaller slices into "All Other Categories" for better readability
threshold = 0.05  # Group categories smaller than 5% of total
total_cases = cases_by_descent.sum()
large_slices = cases_by_descent[cases_by_descent / total_cases > threshold]
small_slices = cases_by_descent[cases_by_descent / total_cases <= threshold]
if not small_slices.empty:
    large_slices['All Other Categories'] = small_slices.sum()

# Simplified pie chart
plt.figure(figsize=(10, 8))
plt.pie(
    large_slices,
    labels=large_slices.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.tab20.colors
)
plt.title('Total Crime Cases by Victim Descent')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.cm import ScalarMappable

# Filter for incidents likely to be school shootings involving firearms
school_shootings = la_crime_data_2010_to_2021_sorted[
    la_crime_data_2010_to_2021_sorted['Premis Desc'].str.contains('SCHOOL', na=False) &
    la_crime_data_2010_to_2021_sorted['Weapon Desc'].str.contains('GUN|FIREARM|RIFLE|PISTOL', na=False, case=False)
]

# Filtering for defined genders 'M', 'F', 'X' (adjust as necessary)
valid_genders = ['M', 'F', 'X']
filtered_data = school_shootings[school_shootings['Vict Sex'].isin(valid_genders)]

# Grouping data by gender and calculating average crime code for normalization
stats_by_gender = filtered_data.groupby('Vict Sex').agg({
    'Crm Cd': 'mean'  # Calculate the average crime code (severity of crime)
}).reset_index()

# Normalizing the severity
stats_by_gender['Normalized Crime Seriousness'] = (stats_by_gender['Crm Cd'] - stats_by_gender['Crm Cd'].min()) / (stats_by_gender['Crm Cd'].max() - stats_by_gender['Crm Cd'].min())

# Mapping the normalized values to the respective genders for the pie chart
normalized_severity = stats_by_gender.set_index('Vict Sex')['Normalized Crime Seriousness']

# Create a colormap from light red to dark red
cmap = LinearSegmentedColormap.from_list("severity", ["#ffcccc", "#ff0000"])
norm = plt.Normalize(min(normalized_severity), max(normalized_severity))

# Creating the pie chart
fig, ax = plt.subplots()
colors = [cmap(norm(value)) for value in normalized_severity]
wedges, texts, autotexts = ax.pie(filtered_data['Vict Sex'].value_counts(), labels=filtered_data['Vict Sex'].value_counts().index, autopct='%1.1f%%', startangle=90, colors=colors)

# Add a colorbar for visual reference of severity
sm = ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, orientation='vertical')
cbar.set_label('Normalized Crime Seriousness')

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Title and visual layout adjustments
plt.title('Percentage of School Shootings by Victim Gender')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter for gun-related crimes
gun_crimes = la_crime_data_2010_to_2021_sorted[
    la_crime_data_2010_to_2021_sorted['Weapon Desc'].str.contains('GUN|FIREARM|RIFLE|PISTOL', na=True, case=False)
]

# Filter for non-gun violent crimes
violent_crimes = la_crime_data_2010_to_2021_sorted[
    la_crime_data_2010_to_2021_sorted['Crm Cd Desc'].str.contains('ASSAULT|ROBBERY|HOMICIDE', na=True, case=False)
]
non_gun_violent_crimes = violent_crimes[
    ~violent_crimes['Weapon Desc'].str.contains('GUN|FIREARM|RIFLE|PISTOL', na=True, case=False)
]

# Average severity for gun and non-gun crimes
average_severity_gun = gun_crimes['Crm Cd'].mean()
average_severity_non_gun = non_gun_violent_crimes['Crm Cd'].mean()

print(f"Average Crime Code for Gun-Related Crimes: {average_severity_gun}")
print(f"Average Crime Code for Non-Gun Violent Crimes: {average_severity_non_gun}")

# Group by month to get counts
gun_crimes_monthly = gun_crimes.resample('ME', on='DATE OCC').size()
non_gun_violent_crimes_monthly = non_gun_violent_crimes.resample('ME', on='DATE OCC').size()

# Plotting
plt.figure(figsize=(14, 7))
gun_crimes_monthly.plot(label='Gun-Related Crimes')
non_gun_violent_crimes_monthly.plot(label='Non-Gun Violent Crimes')
plt.title('Monthly Crime Frequency')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter columns for gun-related crimes
gun_related_crimes = la_crime_data_2010_to_2021_sorted[
    la_crime_data_2010_to_2021_sorted['Weapon Desc'].str.contains("GUN", na=False)
]

# Analyze the types of guns used
guns_used_counts = gun_related_crimes['Weapon Desc'].value_counts()

# Calculate percentage of each gun type
total_gun_crimes = guns_used_counts.sum()
gun_percentages = (guns_used_counts / total_gun_crimes) * 100

# Consolidate categories smaller than 2.5% into "Other"
guns_used_counts = guns_used_counts[gun_percentages >= 2.5]
guns_used_counts["Other"] = total_gun_crimes - guns_used_counts.sum()

# Create a pie chart showing the percentage of different guns used
plt.figure(figsize=(10, 8))
guns_used_counts.plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    labels=guns_used_counts.index,
    title="Percentage of Different Guns Used in Crimes"
)
plt.ylabel("")  # Remove the default y-axis label
plt.show()
```
